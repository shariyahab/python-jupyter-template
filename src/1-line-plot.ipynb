### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.arange(0, 11)
y = x ** 2  # Square of x

# Create the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Customization
plt.title('Square Function Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values (X^2)')
plt.grid(True)
plt.legend()

# Show the line plot
plt.show()

# Create bar plot data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customization
plt.title('Category Value Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the bar plot
plt.show()
```
